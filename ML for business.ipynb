# Introducción

La compañía minera OilyGiant necesita encontrar el mejor lugar para un nuevo pozo de extracción de petróleo. Se nos pide la construcción de un modelo que sea capaz de predecir el volumen de reservas de petróleo en los nuevos pozos y poder elegir la región con mayor beneficio total para los pozos seleccionados.

La compañía nos proporciona datos sobre muestras de petróleo de tres regiones distintas en los que ya se conocer los parámetros de cada pozo de petróleo de la región.

## Estrategia

Para el desarrollo de este proyecto se seguirá la siguiente estrategia:

   1. Los datos se dividirán en dos conjuntos, uno de entrenamiento y uno de validación.
   1. Se entrenará un modelo de tipo regresión lineal para cada región.
   1. Con los resultados obtenidos, calcularemos las ganancias de los pozos.
   1. Escoger los pozos con mayores ganancias de cada región y calcular el beneficio por el volumen de reservas obtenido.
   1. Calcular los riesgos y ganancias para cada región.

## Inicialización

Para el desarrollo del proyecto, primero importamos las librerías necesarias para instanciar el modelo y entrenarlo, métricas para evaluarlo, así como `pandas` para trabajar con conjuntos de datos.

```python
# Carga las librerías

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy import stats as st
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

## Carga los datos

Para este proyecto contamos con tres conjuntos de datos sobre la exploración geológica de tres regiones distintas.

```python
# Carga los datos

try:
    region_data_1 = pd.read_csv('geo_data_0.csv')
    region_data_2 = pd.read_csv('geo_data_1.csv')
    region_data_3 = pd.read_csv('geo_data_2.csv')
    
except:
    region_data_1 = pd.read_csv('/datasets/geo_data_0.csv')
    region_data_2 = pd.read_csv('/datasets/geo_data_1.csv')
    region_data_3 = pd.read_csv('/datasets/geo_data_2.csv')
```

Cada dataframe contiene los siguientes datos:

## Preparar los datos

En este apartado prepararemos los datos para el entrenamiento del modelo, eliminando la columna `id` y segmentando los datasets. 

### Eliminar datos no relevantes

Eliminamos la columna `id` de cada DataFrame para quedarnos con las variables características y objetivo.

```python
# Elimina la columna 'id' de los datasets

region_data_1 = region_data_1.drop('id', axis=1)
region_data_2 = region_data_2.drop('id', axis=1)
region_data_3 = region_data_3.drop('id', axis=1)
```

### Identificar las características y objetivo

Identificamos en cada dataset las características y el objetivo para entrenar un modelo para cada región.

```python
# Identifica las características de cada dataset

features_reg1 = region_data_1.drop('product', axis=1)
features_reg2 = region_data_2.drop('product', axis=1)
features_reg3 = region_data_3.drop('product', axis=1)

# Identifica el objetivo de cada dataset

target_reg1 = region_data_1['product']
target_reg2 = region_data_2['product']
target_reg3 = region_data_3['product']
```

### Segmentar los datos

Para entrenar y evaluar nuestro modelo, segmentamos los datasets en conjuntos de entrenamiento y validación con una relación 75:25 (%)

```python
# Segmenta los datasets en conjunto de entrenamiento y conjunto de validación.

# Segmentación del features & target de la región 1
features_reg1_train, features_reg1_valid, target_reg1_train, target_reg1_valid = train_test_split(
    features_reg1, target_reg1, test_size=0.25, random_state=314
)

# Segmentación del features & target de la región 2
features_reg2_train, features_reg2_valid, target_reg2_train, target_reg2_valid = train_test_split(
    features_reg2, target_reg2, test_size=0.25, random_state=314
)

# Segmentación del features & target de la región 3
features_reg3_train, features_reg3_valid, target_reg3_train, target_reg3_valid = train_test_split(
    features_reg3, target_reg3, test_size=0.25, random_state=314
)
```

```python
# Comprobamos el tamaño de los datasets segmentados de una de las regiones

print('Tamaño total del dataset:', region_data_1.shape)
print()
print('Tamaño de features_reg1_train:', features_reg1_train.shape)
print('Tamaño de features_reg1_valid:', features_reg1_valid.shape)
print()
print('Tamaño de target_reg1_train:', target_reg1_train.shape)
print('Tamaño de target_reg1_valid:', target_reg1_valid.shape)
```

# Entrenar el modelo

En esta parte crearemos un modelo de regresión lineal para cada región y evaluaremos su desempeño con algunas métricas. Utilizaremos un modelo de este tipo ya que la regresión lineal es más adecuada para este caso.

## Modelo de regresión lineal

Identificaremos los modelos de cada región con el sufijo `_regX`, con _X_ siendo el identificador de región.

```python
# Instancia y entrena los modelos para cada región

model_reg1 = LinearRegression().fit(features_reg1_train, target_reg1_train)     # Región 1

model_reg2 = LinearRegression().fit(features_reg2_train, target_reg2_train)     # Región 2

model_reg3 = LinearRegression().fit(features_reg3_train, target_reg3_train)     # Región 3
```

**Predicciones del modelo para cada región**

Guardaremos las predicciones de cada uno de los modelos en las variables `predict_regX`, con _X_ siendo el identificador de región.

```python
# Predicciones de los modelos

predict_reg1 = model_reg1.predict(features_reg1_valid)      # Predicciones región 1

predict_reg2 = model_reg2.predict(features_reg2_valid)      # Predicciones región 2

predict_reg3 = model_reg3.predict(features_reg3_valid)      # Predicciones región 3
```

### Evaluación del modelo

Evualuaremos los modelos calculando el valor _REMC_ y el valor $R^2$ de cada región.



```python
# Muestra el valor promedio de las reservas previstas y el RECM

rmse_reg1 = mean_squared_error(target_reg1_valid, predict_reg1) ** 0.5       # REMC de la región 1
r2_score_reg1 = model_reg1.score(features_reg1_valid, target_reg1_valid)     # Valor R2 de la región 1

rmse_reg2 = mean_squared_error(target_reg2_valid, predict_reg2) ** 0.5       # REMC de la región 2
r2_score_reg2 = model_reg2.score(features_reg2_valid, target_reg2_valid)     # Valor R2 de la región 2

rmse_reg3 = mean_squared_error(target_reg3_valid, predict_reg3) ** 0.5       # REMC de la región 3
r2_score_reg3 = model_reg3.score(features_reg3_valid, target_reg3_valid)     # Valor R2 de la región 3



print(f'Promedio de las reservas previstas en la región 1: {predict_reg1.mean():.3f} miles de barriles')
print('Valor REMC del modelo:', rmse_reg1)
print('Valor R2 del modelo:', r2_score_reg1)
print()
print(f'Promedio de las reservas previstas en la región 2: {predict_reg2.mean():.3f} miles de barriles')
print('Valor REMC del modelo en la región 2:', rmse_reg2)
print('Valor R2 del modelo:', r2_score_reg2)
print()
print(f'Promedio de las reservas previstas en la región 3: {predict_reg3.mean():.3f} miles de barriles')
print('Valor REMC del modelo en la región 3:', rmse_reg3)
print('Valor R2 del modelo:', r2_score_reg3)
```

### Análisis de resultados

1. El promedio de reservas previstas en la región 1 es de 92.777 miles de barriles con un error de 37.728 miles de barriles. Además, el valor $R^2$ es de 0.274, lo que nos indica que la calidad del modelo puede considerarse buena.
<br>

2. En cuanto al modelo de la región 2 es casi perfecto, en dónde obtenemos unas reservas previstas 68.735 miles de barriles en promedio con un error de 0.89 miles de barriles. Se puede apreciar un valor $R^2$ de 0.999, indicándonos un modelo que realiza predicciones con mucha exactitud.
<br>

3. En la región 3 obtenemos un promedio de reservas similar a la región 1. El promedio de reservas previstas en esta región es de 94.946 miles de barriles con un error de 39.956 miles de barriles. Aunque tenemos un valor de $R^2$ algo menor que en la región 1; para este caso sería de 0.195.

### Prueba de cordura

Realizamos la prueba de cordura del modelo comparandolo con un modelo constante que predice el valor objetivo medio para cada observación.

```python
# Modelo constante con valores medios

predict_const1 = pd.Series(target_reg1_train.mean(), index=target_reg1_valid.index)     # Modelo constante para la Región 1
rmse_const1 = mean_squared_error(target_reg1_valid, predict_const1) ** 0.5

predict_const2 = pd.Series(target_reg2_train.mean(), index=target_reg2_valid.index)     # Modelo constante para la Región 2
rmse_const2 = mean_squared_error(target_reg2_valid, predict_const2) ** 0.5

predict_const3 = pd.Series(target_reg3_train.mean(), index=target_reg3_valid.index)     # Modelo constante para la Región 3
rmse_const3 = mean_squared_error(target_reg3_valid, predict_const3) ** 0.5

print('Valor REMC del modelo constante 1:', rmse_const1)
print('Valor REMC del modelo constante 2:', rmse_const2)
print('Valor REMC del modelo constante 3:', rmse_const3)
```

**Conclusiones intermedias**

En las regiones 1 y 3 observamos que el valor _REMC_ difiere a favor de los modelos que hemos entrenado con diferencias de 7000 barriles más en la región 1 y 5000 barriles más de diferencia de error en la región 3, mientras que el modelo constante de la región 2 arroja un promedio de reservas previstas similar al de las otras regiones y nuestro modelo entrenado tiene un error inferior a $\pm$ 1000 barriles

# Cálculo de ganancias

En la siguiente parte del proyecto se tratará el cálculo de las posibles ganancias en promedio que las regiones pueden generar. Se seleccionarán los 200 mejores puntos de cada región para construir los pozos. Se realizará además una evaluación de los riesgos de pérdidas.

## Preparar las variables

Almacenaremos los valores que obtuvimos en el apartado anterior, junto con otros valores que nos proporciona la compañía, en variables necesarias para realizar los siguientes cálculos de ganancias.

_Nota: el presupuesto para el desarrollo de 200 pozos es de 100 millones de dólares. El barril de materias primas genera 4.5 USD de ingresos. EL ingreso de una unidad de producto es de 4500 USD (una unidad de producto son mil barriles)._

```python
# Variables para los cálculos

barrels_unit_income = 4500                      # Ingresos por unidad de producto (mil barriles)
budget = 100000000                              # Presupuesto para 200 pozos (en dólares)
budget_per_well = budget/200                    # Presupuesto por cada pozo (en dólares)

mean_reserves_reg1 = predict_reg1.mean()        # Promedio de reservas previstas en la región 1
mean_reserves_reg2 = predict_reg2.mean()        # Promedio de reservas previstas en la región 2
mean_reserves_reg3 = predict_reg3.mean()        # Promedio de reservas previstas en la región 3

reg1_error = rmse_reg1 * barrels_unit_income    # Margen de error en las ganancias de la región 1
reg2_error = rmse_reg2 * barrels_unit_income    # Margen de error en las ganancias de la región 1
reg3_error = rmse_reg2 * barrels_unit_income    # Margen de error en las ganancias de la región 1

reg1_profitable_wells = pd.Series(predict_reg1).sort_values(ascending=False).iloc[:200]     # Top 200 mejores puntos en la región 1
reg2_profitable_wells = pd.Series(predict_reg2).sort_values(ascending=False).iloc[:200]     # Top 200 mejores puntos en la región 2
reg3_profitable_wells = pd.Series(predict_reg3).sort_values(ascending=False).iloc[:200]     # Top 200 mejores puntos en la región 3
```

## Volumen de reservas mínimo para un pozo rentable

Calcularemos el volumen de reservas mínimo necesario para desarrollar un pozo sin que produzca pérdidas. Conocemos los ingresos que se generan por unidad de producto y el presupuesto para el desarrollo de un nuevo pozo. 

Necesitamos un volumen de reservas tal que $($`min_reserves` $\times$ `barrels_unit_income`$)$ $-$ `budget_per_well` $=$ 0.

```python
# Volumen mínimo de reservas para un pozo rentable

min_reserves = budget_per_well / barrels_unit_income

print(f'El volumen mínimo de reservas necesarias para desarrollar un nuevo pozo sin pérdidas es de: {min_reserves:.3f} miles de barriles')
```

Sería necesario producir más de 111 mil barriles para que un pozo sea rentable.

Comparamos este resultado con los volúmenes promedio previstos por los modelos para las distintas regiones.

```python
reg1_revenue = (mean_reserves_reg1 * barrels_unit_income) - budget_per_well     # Ganancias según promedio de reservas en la región 1
reg2_revenue = (mean_reserves_reg2 * barrels_unit_income) - budget_per_well     # Ganancias según promedio de reservas en la región 2
reg3_revenue = (mean_reserves_reg3 * barrels_unit_income) - budget_per_well     # Ganancias según promedio de reservas en la región 3

print('Ganancias previstas según el promedio de reservas en la región 1: de',
        round((reg1_revenue - reg1_error), 2), 'USD a', round((reg1_revenue + reg1_error), 2), 'USD')
print()
print('Ganancias previstas según el promedio de reservas en la región 2: de',
        round((reg2_revenue - reg2_error), 2), 'USD a', round((reg2_revenue + reg2_error), 2), 'USD')
print()
print('Ganancias previstas según el promedio de reservas en la región 3: de',
        round((reg3_revenue - reg3_error), 2), 'USD a', round((reg3_revenue + reg3_error), 2), 'USD')
```

### Gráfica de ganancias según el promedio previsto

Realizamos una representación gráfica de los resultados obtenidos para tener una idea clara de los ingresos estimados según la media de reservas previstas en cada región.

```python
# Crea un dataframe con los resultados anteriores

data_graph = [
    ['Región 1', reg1_revenue - reg1_error, reg1_revenue, reg1_revenue + reg1_error],
    ['Región 2', reg2_revenue - reg2_error, reg2_revenue, reg2_revenue + reg2_error],
    ['Región 3', reg3_revenue - reg3_error, reg3_revenue, reg3_revenue + reg3_error]
]
column = ['region', 'min_mean_revenue', 'mean_revenue', 'max_mean_revenue']

revenue_graph = pd.DataFrame(data_graph, columns=column)


# Imprime un gráfico que mustra los ingresos según el promedio obtenido en cada región teniendo en cuenta los valores de REMC

revenue_graph.plot(kind='bar', x='region', figsize=(12, 6))

plt.style.use('seaborn')
plt.grid(visible=True, linestyle='--', linewidth=1, alpha=0.8)

plt.legend(labels=['Ingreso medio mínimo', 'Ingreso medio', 'Ingreso medio máximo'])
plt.title('Ingresos estimados por región (en USD)')

plt.show()
```

## Selección de la región más rentable

### Función para el cálculo de beneficio

A continuación definiremos una función para calcular los beneficios máximos que se pueden obtener de cada región. Dicha función hará lo siguiente:

   1. Escogerá los 200 puntos con mayores reservas según la predicción.
   1. Sumará el volumen total de reservas de la región.
   1. Calculará el beneficio por el volumen de reservas obtenido.

```python
# Define una función para calcular el máximo beneficio de una región

def calculate_profit(predictions, budget, income):

    max_predictions = pd.Series(predictions).sort_values(ascending=False).iloc[:200]
    total_volume = max_predictions.sum()

    total_income = total_volume * income
    profit = total_income - budget

    return profit
```

### Región más rentable

Con la función que hemos definido, haremos un bucle que realice el cálculo para las 3 regiones y seleccione aquella con mayor beneficio según las predicciones. Después se mostrará el beneficio que se podría obtener del resto de regiones.

```python
# Compara los beneficios de las distintas regiones y obtiene la región con mayores beneficios

reserves_predictions = [predict_reg1, predict_reg2, predict_reg3]
profit = 0
max_profit = 0

for index, prediction in enumerate(reserves_predictions):

    profit = calculate_profit(prediction, budget=budget, income=barrels_unit_income)

    if profit > max_profit:
       max_profit = profit
       region = index + 1
    

print('La región con mayores beneficios es la Región {} con unos beneficios totales de \
{:.3f} millones de dólares'.format(region, max_profit/1000000))
```

Hemos podido comprobar que, seleccionando los 200 mejores puntos para el desarrollo de nuevos pozos, en la Región 1 es dónde se obtendrían mayores beneficios netos.

Los beneficios que se podrían obtener en el resto de regiones son los siguientes:

# Cálculo de riesgos y ganancias

Ya hemos podido observar los posibles beneficios máximos que se podrían obtener de cada región. El siguiente paso será calcular los riesgos y ganancias de cada región para encontrar aquella región que nos pudiera proporcionar más ganancias minimizando el riesgo.

## Bootstrapping

Para este procedimiento, aplicaremos la técnica de _bootstrapping_ con 1000 muestras. En cada muestra escogeremos al azar 500 ubicaciones de cada región, seleccionaremos las 200 mejores y calculamos el promedio de reservas de cada muestra.

```python
# Tecnica de bootsrtapping

bootstrapped_samples = [predict_reg1, predict_reg2, predict_reg3]


# Recorre la lista con las predicciones
for i, region in enumerate(bootstrapped_samples):
    
    region_reserves = []    # Lista con las medias del bootstrapping

    # Bootstrapping con 1000 muestras
    for j in range(1000):

        # Coge una muestra de 500 ubicaciones al azar y selecciona las 200 mejores
        well_samples = pd.Series(np.random.choice(region, size=500)).sort_values(ascending=False)[:200]
        reserves_mean = well_samples.mean()
        region_reserves.append(reserves_mean)

    bootstrapped_samples[i] = region_reserves
    
```

## Distribución de beneficios de las submuestras

De las submuestras que obtuvimos mediante la técnica de _bootstrapping_ buscamos la distribución del beneficio en cada región.

```python
# Calcula el beneficio por pozo de las submuestras que hemos tomado

bootstrapped_profits = []

for i, region in enumerate(bootstrapped_samples):
    sample_profits = []
    
    for j, subsample in enumerate(region):
        subsample_profit = ((subsample * barrels_unit_income) - budget_per_well) / 1000
        sample_profits.append(subsample_profit)
        
    bootstrapped_profits.append(sample_profits)
```

Y mostramos de forma gráfica como se distribuyen los beneficios netos en cada región de acuerdo a las submuestras.

```python
# Muestra una gráfica con la distribución de beneficios

for i, subsamples in enumerate(bootstrapped_profits):

    sns.histplot(data=bootstrapped_profits[i], bins=30, kde=True, color=color[i], alpha=0.5, palette='muted')
    # Los colores están en una lista definida antes

plt.title('Distribución de beneficios promedio por región (Bootstrapped data)')
plt.xlabel('Beneficios (en miles de USD)')
plt.ylabel('Número de ubicaciones')
plt.legend(title='Leyenda', labels=['Beneficios en la Región 1', 'Beneficios en la Región 2', 'Beneficios en la Región 3'],
            facecolor='white', frameon=True)
plt.show()
```

## Beneficio promedio, intervalo de confianza y riesgo de pérdida

El siguiente paso es calcular el beneficio promedio en un intervalo de confianza del 95% y el porcentaje de riesgo de generar pérdidas en cada región.

```python
# Calcula el beneficio promedio, el intervalo de confianza de cada región y el risgo de pérdidas en tanto por ciento

for i , region in enumerate(bootstrapped_profits):
    region = pd.Series(region)
    
    # Calcula el intervalo de confianza del 95%
    confidence_interval = st.t.interval(
        0.95, len(region-1), region.mean(), region.sem()
    )

    # Calcula el riesgo de pérdidas

    z_score = (0 - region.mean()) / region.var() ** 0.5     # Obtiene Z estandarizando el valor de 0
    losses_risk = st.norm.cdf(z_score) * 100                # Calcula la función de distribución acumulada para los valores igual o menores que 0

    print('El beneficio promedio en la Región {} es {:.3f} miles de dólares por pozo'.format(i+1, region.mean()))
    print('El intervalo de confianza es {}'.format(confidence_interval))
    print('Hay una probabilidad de generar pérdidas del {:.2f} %'.format(losses_risk))
    print()
```

**Observaciones**

Tras este análisis, podemos sugerir desarrollar los pozos en la Región 2, ya que esta región aportaría mayores beneficios en promedio por cada pozo desarrollado, en cambio, tiene mayor probabilidad de pérdidas.
